- (4, 'TX_AMOUNT', 'REAL', 0, None, 0) - сумма транзакции

Также необходимо скачать второй CSV-файл - «client_info.csv», создать таблицу customer_bd со всеми полями, загрузить данные из файла в таблицу и получить таблицу со структурой:

- (0, 'CLIENT_ID', 'INTEGER', 0, None, 0) - id клиента
- (1, 'START_DT', 'NUMERIC', 0, None, 0) - дата начало записи о клиенте
- (2, 'END_DT', 'NUMERIC', 0, None, 0) - дата закрытия записи о клиенте

- (4, 'YEAR_BIRTH', 'TEXT', 0, None, 0) - дата рождение клиента

Примечание для следующих шагов: сумма транзакций не может быть null-ом, это всегда число.

```python
import sqlite3 
from sqlite3 import Error 
import pandas as pd
```

```python
def sql_connection():     
    try:         
        con = sqlite3.connect('HW_3_db.db')         
        return con     
    except Error:         
        print(Error) 
```

```python
con =  sql_connection()
```

# Шаг 2
(Балл - 1 за каждый пункт) Написать скрипты:

- a. Найти имя клиента/клиентов с максимальной суммой транзакций за весь период (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого мы считаемся).
- b. Найти имя клиента/клиентов с минимальной суммой транзакций за весь период (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого мы считаемся).
- c. Найти сумму транзакций относительно даты 2023-04-01 для клиентов, у которых id начинается с 111 (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого мы считаемся). Транзакции учитываются только после завершении дня.
- d. Найти сумму транзакций относительно года рождения клиентов (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого мы считаемся). Сортировать по убыванию года рождения.
- e. Найти количество транзакций для каждого клиента (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого мы считаемся). Сортировать по убыванию количества транзакций.

```python
# За дефолтную дату для выборки действующих клиентов брать ‘2023-05-01’
default_date = '2023-05-01'
```

```python
# a. Найти имя клиента/клиентов с максимальной суммой транзакций за весь период (клиент должен быть действующим, то есть 
# дата закрытия записи о клиенте не должна быть меньше дня относительно которого мы считаемся).
query_1  = f"""
SELECT CLIENT_NAME,  max(SUM_TRANS) 

FROM (SELECT CLIENT_NAME, sum(TX_AMOUNT) SUM_TRANS, END_DT, date(TX_DATETIME)
      FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
      WHERE END_DT >= date('{default_date}')
      GROUP BY CLIENT_ID);
"""
df1 = pd.read_sql(query_1, con)
df1
```

```python
# b. Найти имя клиента/клиентов с минимальной суммой транзакций за весь период (клиент должен быть действующим, то есть
# дата закрытия записи о клиенте не должна быть меньше дня относительно которого мы считаемся).
query_2  = f"""
SELECT CLIENT_NAME,  min(SUM_TRANS) 

FROM (SELECT CLIENT_NAME, sum(TX_AMOUNT) SUM_TRANS, END_DT, date(TX_DATETIME)
      FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
      WHERE END_DT >= date('{default_date}')
      GROUP BY CLIENT_ID);
"""
df2 = pd.read_sql(query_2, con)
df2
```

```python
# c. Найти сумму транзакций относительно даты 2023-04-01 для клиентов, у которых id начинается с 111 (клиент должен быть
# действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого мы считаемся). 
# Транзакции учитываются только после завершении дня.

# Комментарий преподавателя:
# Правильный вариант - тут требуется группировка по клиентам.
# Давайте будем считать что относительно даты 2023-04-01 имеется ввиду транзакции до этой даты (для корректного вывода ответа),
# то есть мы должны по клиентам которые действительны к моменту 2023-04-01 найти сумму их транзакций за весь период
# до 2023-04-01 (не включая его, так как в условии было указано что учитываются транзакции только после завершения дня,
# значит день относительно которого мы считаемся не должен учитываться)

query_3  = """
SELECT CLIENT_ID, sum(TX_AMOUNT) SUM_TRANS
FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
WHERE CLIENT_ID LIKE '111%' AND END_DT >= '2023-04-01' AND date(TX_DATETIME) < '2023-04-01'
GROUP BY CLIENT_ID;
"""
df3 = pd.read_sql(query_3, con)
df3
```

```python
query_3  = """
select
        t.CUSTOMER_ID,
        sum(t.TX_AMOUNT) as TOTAL_AMOUNT
    from
        transaction_bd t
        join customer_bd c on t.CUSTOMER_ID = c.CLIENT_ID
    where c.END_DT >= DATE('2023-04-01')
        and t.TX_DATETIME < DATE('2023-04-01')
        and t.CUSTOMER_ID like '111%'
    group by t.CUSTOMER_ID
    order by t.CUSTOMER_ID
;"""
df3 = pd.read_sql(query_3, con)
df3

```

```python
# d. Найти сумму транзакций относительно года рождения клиентов (клиент должен быть действующим, то есть дата закрытия 
# записи о клиенте не должна быть меньше дня относительно которого мы считаемся). Сортировать по убыванию года рождения.

# Комментарий преподавателя:
# Правильный вариант - сгрупировать просто по году рождения клиентов, то есть на выходе должны иметь год
# рождения и сумму транзакций

query_4  = f"""
SELECT YEAR_BIRTH, sum(TX_AMOUNT) SUM_TRANS
FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
WHERE END_DT >= date('{default_date}')
GROUP BY YEAR_BIRTH
ORDER BY YEAR_BIRTH DESC;
"""
df4 = pd.read_sql(query_4, con)
df4
```

```python
# e. Найти количество транзакций для каждого клиента (клиент должен быть действующим, то есть дата закрытия записи о 
# клиенте не должна быть меньше дня относительно которого мы считаемся). Сортировать по убыванию количества транзакций.
query_5  = f"""
SELECT CLIENT_NAME, CLIENT_ID, count(TX_AMOUNT)
FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
WHERE END_DT >= date('{default_date}')
GROUP BY CLIENT_ID
ORDER BY count(TX_AMOUNT) DESC;
"""
df5 = pd.read_sql(query_5, con)
df5
```

```python
query_6  = f"""
SELECT strftime('%m',TX_DATETIME) MONTH, sum(TX_AMOUNT)
FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
WHERE END_DT >= date('{default_date}')
GROUP BY strftime('%m',TX_DATETIME);
"""
df6 = pd.read_sql(query_6, con)
df6
```
