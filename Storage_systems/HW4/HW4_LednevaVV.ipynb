- e. Вычислить интегральное распределение суммы транзакций для каждого дня за весь период, то есть вывести дату транзакции, сумму транзакции и интегральное распределение по сумме (не учитываем что клиент может быть недействующим)
 

```python
# За дефолтную дату для выборки действующих клиентов брать ‘2023-05-01’
default_date = '2023-05-01'
```

```python
# a. Вывести список всех клиентов, у которых между двумя ближайшими транзакциями был перерыв больше 35 дней хотя бы один раз
# (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого 
# мы считаемся).
query_1  = f"""
SELECT  DISTINCT 
    CLIENT_NAME,
    CLIENT_ID
FROM(    
    SELECT CLIENT_NAME,
    CLIENT_ID,
    (strftime('%s',"TX_DATETIME")-(LAG(strftime('%s',"TX_DATETIME")) OVER(PARTITION BY CUSTOMER_ID ORDER BY TX_DATETIME))) DIF_sec 
    FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
    WHERE END_DT >= date('{default_date}'))
WHERE DIF_sec>(60*60*24*35);
"""
df1 = pd.read_sql(query_1, con)
df1
```

```python
# b. Вывести список клиентов, у которых максимальная сумма транзакции (имеется ввиду максимальное значение TX_AMOUNT) 
# больше 50000. В ответе должен быть id клиента, максимальное значение суммы транзакций, минимальное значение суммы транзакций 
# (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого 
# мы считаемся).
query_2  = f"""
SELECT DISTINCT
    CLIENT_ID,
    max_TX,
    min_TX
FROM
    (SELECT 
    CLIENT_ID, 
    max(TX_AMOUNT) OVER(PARTITION BY CLIENT_ID) max_TX,
    min(TX_AMOUNT) OVER(PARTITION BY CLIENT_ID) min_TX
    FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
    WHERE END_DT >= date('{default_date}'))
WHERE max_TX>50000;
"""
df2 = pd.read_sql(query_2, con)
df2
```

```python
# c. Вывести топ-10 клиентов (сортируя их по убыванию id), у которых сумма дневных транзакций больше суммы вечерних транзакций
# Условимся, что дневными считаются транзакции, которые были произведены до 12 часов, а вечерними - после 12 часов включительно.
# (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня относительно которого
# мы считаемся).
query_3  = f"""
SELECT DISTINCT
    CLIENT_NAME,
    CLIENT_ID
FROM
    (SELECT 
    CLIENT_NAME,
    CLIENT_ID,
    (sum(TX_AMOUNT) FILTER(WHERE time(TX_DATETIME) < time('12:00:00')) OVER(PARTITION BY CLIENT_ID))-(sum(TX_AMOUNT) FILTER(WHERE time(TX_DATETIME) >= time('12:00:00')) OVER(PARTITION BY CLIENT_ID)) DIF
    FROM transaction_bd JOIN customer_bd ON CUSTOMER_ID = CLIENT_ID
    WHERE END_DT >= date('{default_date}'))
WHERE DIF>0
ORDER BY CLIENT_ID DESC
LIMIT 10;
"""
df3 = pd.read_sql(query_3, con)
df3
```

```python
# d. Найти для каждого клиента день, когда у него была максимальная сумма транзакции, то есть вывести клиента, дату транзакции,
# сумму транзакции (клиент должен быть действующим, то есть дата закрытия записи о клиенте не должна быть меньше дня 
# относительно которого мы считаемся).
query_4  = f"""
SELECT 
    CLIENT_ID, 
    TX_DATETIME,
    max_TX
FROM
    (SELECT 
    CLIENT_ID, 
    TX_DATETIME,
    TX_AMOUNT,
    max(TX_AMOUNT) OVER(PARTITION BY CLIENT_ID) max_TX
    FROM customer_bd LEFT JOIN transaction_bd  ON CLIENT_ID = CUSTOMER_ID
    WHERE END_DT >= date('{default_date}'))
WHERE max_TX = TX_AMOUNT

;
"""
df4 = pd.read_sql(query_4, con)
df4
```

```python
# e. Вычислить интегральное распределение суммы транзакций для каждого дня за весь период, то есть вывести дату транзакции,
# сумму транзакции и интегральное распределение по сумме (не учитываем что клиент может быть недействующим)


# Не однозначная формулировка в задании: "Вычислить интегральное распределение суммы транзакций для каждого дня за весь период"
# Не понятно, вычислить распределение по транзакциям внутри каждого дня за весь период или все такие сгрупировать по дням и
# считать вклад общей суммы за каждый день за весь период

# Вариант №1. Распределение внутри каждого дня

query_5  = f"""
SELECT  
    date(TX_DATETIME),
    TX_AMOUNT,
    CUME_DIST() OVER(PARTITION BY date(TX_DATETIME) ORDER BY TX_AMOUNT) CUME_DIST_TX
FROM transaction_bd;
"""
df5 = pd.read_sql(query_5, con)
df5
```

```python
# e. Вариант №2. Распределение суммы за каждый день за весь период
query_6  = f"""
SELECT  
    DATE,
    TX_AMOUNT,
    CUME_DIST() OVER(ORDER BY TX_AMOUNT) CUME_DIST_TX
FROM 
    (SELECT
    date(TX_DATETIME) DATE,
    SUM(TX_AMOUNT) TX_AMOUNT
    FROM transaction_bd 
    GROUP BY date(TX_DATETIME));
"""
df6 = pd.read_sql(query_6, con)
df6
```
