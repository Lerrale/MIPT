### key_skills

Extract key skills from the key_skills column.

Извлечем ключевые навыки из столюца key_skills.

```python
df = pd.read_csv("reached_data2.csv")
```

```python
# Replace gaps with an empty list.
#
# Заменяем пропуски на пустой список.
def empty_assign(s):
    idx = s.isna()
    s.loc[idx] = [[]] * idx.sum()
    
empty_assign(df['key_skills'])
```

```python
# Replace json with a list of skills.
#
# Заменяем json на список скилов.
from json import loads
def extract_key_skills(text):
    #print(text)
    #text = text.to_json(orient='records', force_ascii=False)
    ##text=loads(text)
    #text=text[0]
    let_list = []
    #print(text)
    for i in range(len(text)):
        if text[i] not in "[{',:}]":
            let_list.append(text[i])
    let_list = let_list[5:]
    let_list = "".join(let_list)
    let_list = let_list.lower()
    let_list = let_list.split(' name ')
    return let_list
```

```python
df.key_skills = df.key_skills.apply(extract_key_skills)
```

```python
# Using the get_dummies() method to encode skills.
#
# Применение метода get_dummies() для кодирования навыков.
skills_encoded = pd.get_dummies(df['key_skills'].apply(pd.Series).stack()).sum(level=0)
```

```python
skills_encoded
```

We see that there are a huge number of columns with almost the same names.

Мы видим, что есть огромное количество столбцов с почти одинаковыми названиями.

```python
skills_encoded = skills_encoded.iloc[:,1:]
```

```python
def process_similar_columns(df, list_similar_columns):
    for i in range(1, len(list_similar_columns)):
        df[list_similar_columns[0]] = df[list_similar_columns[0]]+df[list_similar_columns[i]]
    df.drop(list_similar_columns[1:], axis=1, inplace=True)
    df[list_similar_columns[0]] = df[list_similar_columns[0]].apply(lambda x: 1 if x>1 else x)         
```

Let's combine the main features.

Объединим основные признаки.

```python
list_similar_column_types = [machine_learning,
                             ms_sql,
                             data_analysys,
                             databases,
                             a_b_test,
                             one_c,
                             agile,
                             airflow,
                             jira,
                             backend,
                             bigdata,
                             bi,
                             catboost,
                             computer_vision,
                             data_engineering,
                             data_science,
                             etl,
                             jupyter_notebook,
                             nlp, 
                             numpy, 
                             postgree_sql,
                             neural_networks]
```

```python
for i in list_similar_column_types:
    process_similar_columns(skills_encoded, i)
```
